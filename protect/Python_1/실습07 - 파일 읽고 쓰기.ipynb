# 실습 #3: 워드 클라우드

1) 인터넷에서 좋은 글(영문으로 된 연설문, 혹은 기사, 책 내용 등)을 찾아, txt로 저장하고 읽어와 워드 클라우드를 만들어 보세요.

- 코드는 예제파일을 참조하세요.
- 불필요한 단어/조사는 제거하세요.
- 글의 내용을 한눈에 파악할 수 있는 멋진 워드클라우드를 만들어 보세요.

```python
# 패키지 설치합
!pip install wordcloud
from pathlib import Path
import numpy as np
from PIL import Image
```

```python
# 파일 읽기
file = open('Dream.txt', 'r', encoding='UTF-8')
text = file.read()
file.close()

# 공백을 구분자로 하여 단어 단위로 자르기
wordList = text.split()
```

```python
# 중복 단어 제거
worduniq = set(wordList)

# 딕셔너리 선언
wordCount = {}

# 단어별 개수 저장
for w in worduniq:
    wordCount[w] = wordList.count(w)

# 제외 대상 조사
del_word = ['the','a','is','are', 'not','of','on','that','this','and','be','to', 'from']

# 제외하기
for w in del_word:
    if w in wordCount:
        del wordCount[w]
```

```python
# 라이브러리 불러오기
import matplotlib.pyplot as plt
from wordcloud import WordCloud
%config InlineBackend.figure_format='retina'

# 워드 클라우드 만들기
wordcloud = WordCloud(font_path = 'C:/Windows/fonts/HMKMRHD.TTF',  # 글꼴
                      width=2000,  # 너비
                      height=1000, # 높이
                      background_color='white').generate_from_frequencies(wordCount) 

# 표시하기
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
