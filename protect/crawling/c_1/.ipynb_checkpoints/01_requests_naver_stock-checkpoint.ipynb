## Web Crawling
- 웹 페이지에서 데이터를 수집하는 방법에 대해서 학습

### 웹크롤링 방법

#### 웹페이지의 종류
- 정적인 페이지 : 웹 브라우져에 화면이 한번 뜨면 이벤트에 의한 화면의 변경이 없는 페이지
- 동적인 페이지 : 웹 브라우져에 화면이 뜨고 이벤트가 발생하면 서버에서 데이터를 가져와 화면을 변경하는 페이지

#### requests 이용
- 받아오는 문자열에 따라 두가지 방법으로 구분
    - json 문자열로 받아서 파싱하는 방법 : 주로 동적 페이지 크롤링할때 사용
    - html 문자열로 받아서 파싱하는 방법 : 주로 정적 페이지 크롤링할때 사용
        
#### selenium 이용
- 브라우져를 직접 열어서 데이터를 받는 방법

#### 크롤링 방법에 따른 속도
- requests json > requests html > selenium

### Crawling Naver Stock Data
- 네이버 증권 사이트에서 주가 데이터 수집
- 수집할 데이터 : 일별 kospi, kosdaq 주가, 일별 환율(exchange rate) 데이터
- 데이터 수집 절차
    - 웹서비스 분석 : url
    - 서버에 데이터 요청 : request(url) > response : json(str)
    - 서버에서 받은 데이터 파싱(데이터 형태를 변경) : json(str) > list, dict > DataFrame

```python
import warnings
warnings.filterwarnings('ignore') # 경고 문구 안뜨게 해주는 설정
import requests
import pandas as pd
```

#### 1. 웹서비스 분석 : url
- pc 웹페이지가 복잡하면 mobile 웹페이지에서 수집

```python
page_size, page = 30, 1
url = f'https://m.stock.naver.com/api/index/KOSPI/price?pageSize={page_size}&page={page}'
url
```

#### 2. 서버에 데이터 요청 : request(url) > response : json(str)
- response의 status code가 200이 나오는지 확인
- 403이나 500이 나오면 request가 잘못되거나 web server에서 수집이 안되도록 설정이 된것임
    - header 설정 또는 selenium 사용
- 200이 나오더라도 response 안에 있는 내용을 확인 > 확인하는 방법 : response.text

```python
response = requests.get(url)
response
```

```python
response.text[:300]
```

#### 3. 서버에서 받은 데이터 파싱(데이터 형태를 변경) : json(str) > list, dict > DataFrame

```python
data = response.json()
```

```python
cols = ['localTradedAt', 'closePrice']
df= pd.DataFrame(data)[cols]
```

```python
df
```

#### 4. 함수로 만들기

```python
def kor_stock(page_size=30, page=1, code='KOSPI'):
    url = url = f'https://m.stock.naver.com/api/index/{code}/price?pageSize={page_size}&page={page}'
    response = requests.get(url)
    data = response.json()
    cols = ['localTradedAt', 'closePrice']
    df = pd.DataFrame(data)[cols]
    
    return df
```

```python
kor_stock(30, 1)
```

```python
kor_stock(30, 1, 'KOSDAQ')
```

#### 5. 원달러 환율 데이터 수집 : 실습

```python
# 1. URL
code, page, page_size= 'FX_USDKRW', 1, 60
url = f'https://m.stock.naver.com/front-api/v1/marketIndex/prices\
?category=exchange&reutersCode={code}&page={page}&pageSize={page_size}'

# 2. respones
response = requests.get(url)
data = response.json()['result']
df_us = pd.DataFrame(data)[['localTradedAt', 'closePrice']]
df_us
```

#### 6. 시각화

```python
%matplotlib inline
%config InlineBackend.figure_formats = {'png', 'retina'}
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
kospi = kor_stock(60, 1, 'KOSPI')
kosdaq = kor_stock(60, 1, 'KOSDAQ')
```

```python
df_us.tail(2)
```

```python
kospi.tail(2)
```

```python
kosdaq.tail(2)
```

```python
korea_stock = kospi.copy()
korea_stock.columns = ['Date','KOSPI']
korea_stock['KOSDAQ'] = kosdaq['closePrice'].apply(lambda data : float(data.replace(',','')))
korea_stock['US_Dollar'] = df_us['closePrice'].apply(lambda data : float(data.replace(',','')))
korea_stock['KOSPI'] = kospi['closePrice'].apply(lambda data : float(data.replace(',','')))
korea_stock.tail(2)
```

```python
dfs = korea_stock[::-1]

plt.figure(figsize=(12, 8))
plt.plot(dfs['Date'], dfs['KOSPI'], label='KOSPI')
plt.plot(dfs['Date'], dfs['KOSDAQ'], label='KOSDAQ')
plt.plot(dfs['Date'], dfs['US_Dollar'], label='US_Dollar')
plt.xticks(dfs['Date'][::4], rotation=45)
plt.legend()  # 범례

plt.show()
```

#### 7. 데이터 스케일링
- min max scaling


- $z = \frac{x_i - min(x)}{max(x) - min(x)} (0 \leqq z \leqq 1)$


- latex syntax : `https://jjycjnmath.tistory.com/117`

```python
from sklearn.preprocessing import minmax_scale
```

```python
# 시각화
plt.figure(figsize=(20, 5))

columns = korea_stock.columns[1:]
for column in columns:
    plt.plot(korea_stock["Date"], minmax_scale(korea_stock[column]), label=column)
    
xticks_count = 11
plt.xticks(korea_stock["Date"][::int(len(korea_stock) // xticks_count) + 1])
plt.legend(loc=0)
plt.show()
```

#### 8. 상관관계 분석
- 피어슨 상관계수(Pearson Correlation Coefficient)
- 두 데이터 집합의 상관도를 분석할때 사용되는 지표
- 상관계수의 해석
    - -1에 가까울수록 서로 반대방향으로 움직임
    - 1에 가까울수록 서로 같은방향으로 움직임
    - 0에 가까울수록 두 데이터는 관계가 없음

```python
corr_df = korea_stock[korea_stock.columns[1:]].corr()
corr_df
```
